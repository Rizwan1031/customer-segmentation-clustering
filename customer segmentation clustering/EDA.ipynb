```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
gg = pd.read_csv('marketing_campaign_cleaned.csv')
gg
```

```python
gg = gg.sort_values(by ='ID' , ascending = True)
gg
```

```python
plt.figure(figsize=(10,6))
sns.histplot(gg['Income'], bins=30 , kde= True)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(gg['Marital_Status'] , gg['Income'])
plt.title('Income by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.scatterplot(x=gg['Income'], y=gg['Kidhome'])
plt.title('Income vs Kidhome')
plt.xlabel('Income')
plt.ylabel('Kidhome')
plt.show()
```

```python
Education_counts = gg['Education'].value_counts()
```

```python
plt.figure(figsize=(7,6))
sns.scatterplot(x=gg['Income'], y=gg['Teenhome'])
plt.title('Income vs Teenhome')
plt.xlabel('Income')
plt.ylabel('Teenhome')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(gg.corr(numeric_only=True) , annot =True , cmap = 'coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
