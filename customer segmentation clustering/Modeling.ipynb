```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd
```

```python
X = pd.read_csv('marketing_campaign_cleaned.csv')
X
```

```python
X_numeric = X.select_dtypes(include=['int64' , 'float64'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)
```

```python
wcss = []
for i in range(1 ,11):
    KM = KMeans(n_clusters = i , random_state = 42)
    KM.fit(X_scaled)
    wcss.append(KM.inertia_)
```

```python
kmeans = KMeans(n_clusters= 5 , random_state=42)
X['Clusters'] = kmeans.fit_predict(X_scaled)

kmeans_2 = KMeans(n_clusters=5, random_state=42)
kmeans_2.fit(X_numeric[['Income', 'Recency']])
X['Clusters_Income_Recency'] = kmeans_2.predict(X_numeric[['Income', 'Recency']])
```

```python
score = silhouette_score(X_scaled , X['Clusters'])
print(score)
print(f"Silhouette Score:{score:.3f}")
```

```python
import pickle
with open('modeling_data.pkl', 'wb') as f:
    pickle.dump({
        'wcss': wcss,
        'X_scaled': X_scaled,
        'X_numeric': X_numeric,
        'X': X,
        'kmeans': kmeans,
        'kmeans_2': kmeans_2,
        'scaler': scaler
    }, f)

print("Data saved successfully!")
```
